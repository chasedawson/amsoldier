```python
df = df.sort_values(by=['classification_id'])
df.head()
```

```python
ids = np.unique(df["classification_id"])
ids
```

```python
df[df['classification_id'] == 77461209]
```

```python
new_tasks = []
for cid in ids:
    tasks = df[df['classification_id'] == cid]['task']
    n_tasks = len(tasks)
    for i in range(n_tasks):
        new_tasks.append(f"T{i}")
```

```python
df['task_renamed'] = new_tasks
```
