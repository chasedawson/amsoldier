```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv('the-american-soldier-classifications-final.csv')
```

```python
df.head()
```

```python
# what is the maximum number of tasks for one image
max_tasks = 0
for i in df.index:
    json_annotation = json.loads(df.annotations[i])
    n_tasks = len(json_annotation)
    if n_tasks > max_tasks:
        max_tasks = n_tasks
max_tasks
```

```python
annotations_cols = []
for i in range(18):
    task = 'T' + str(i)
    annotations_cols += [task, task + "_label", task + "_value"]
annotations_cols
```

```python
colnames = ['key', 'classification_id', 'user_name', 'subject_ids', 'image_id', 'workflow_name'] + annotations_cols
colnames
```

```python
res_df = pd.DataFrame(index=range(len(df)), columns=colnames)
res_df
```

```python
res_df.key = df.key.copy()
res_df.classification_id = df.classification_id.copy()
res_df.user_name = df.user_name.copy()
res_df.subject_ids = df.subject_ids.copy()
res_df.image_id = df.image_id.copy()
res_df.workflow_name = df.workflow_name.copy()
```

```python
for i in res_df.index:
    json_annotations = json.loads(df.annotations[i])
    for j in range(len(json_annotations)):
        task = 'T' + str(j)
        res_df.at[i, task] = json_annotations[j]["task"] if "task" in json_annotations[j] else ""
        res_df.at[i, task + "_label"] = json_annotations[j]["task_label"] if "task_label" in json_annotations[j] else ""
        res_df.at[i, task + "_value"] = json_annotations[j]["value"] if "value" in json_annotations[j] else ""
```

```python
res_df = res_df.fillna("<none>")
res_df
```

```python
import re
```

```python
# get all value columns
value_cols = ['T' + str(x) + '_value' for x in range(18)]
value_cols
```

```python
for col in value_cols:
    res_df[col] = res_df[col].astype(str)
```

```python
res_df.info()
```

```python
p_regex_pattern = "[(\{)|(\[)][p|P][a|A][r|R][a|A][g|G][r|R][a|A][p|P][h|H][(\})|(\])]"
```

```python
p_end_regex_pattern = "[(\{)|(\[)][(\\\\)^(\/)][p|P][a|A][r|R][a|A][g|G][r|R][a|A][p|P][h|H][(\})|(\])]"
```

```python
def count_paragraphs(data):
    p_counts = 0
    p_end_counts = 0
    for index, row in data.iterrows():
        for col in value_cols:
            p_locs = [m.start() for m in re.finditer(p_regex_pattern, str(row[col]))]
            p_end_locs = [m.start() for m in re.finditer(p_end_regex_pattern, str(row[col]))]
            p_counts += len(p_locs)
            p_end_counts += len(p_end_locs)
    return (p_counts, p_end_counts)
```

```python
p_counts1 = count_paragraphs(res_df)
```

```python
# replace all occurrences of [paragraph] tags with \n
for index, row in res_df.iterrows():
    for col in value_cols:
        # replace all occurrences of [paragraph] tags with \n
        res_df.at[index, col] = re.sub(p_regex_pattern, "\n\n", res_df.at[index, col])

        # remove all occurrences of [/paragraph]
        res_df.at[index, col] = re.sub(p_end_regex_pattern, "", res_df.at[index, col])

        # convert all instances of two or more \n next to each other (regardless of spacing) to two side by side \n
        res_df.at[index, col] = re.sub("(\s*\n\s*)(\s*\n\s*)+", "\n\n", res_df.at[index, col])
res_df
```
