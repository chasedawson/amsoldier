```python
import pandas as pd
import numpy as np
import json
import math

classifications = pandas.read_csv('the-american-soldier-classifications-final.csv')
```

```python
classifications.head()
```

```python
tasks = []
for i in classifications.index:
    json_annotation = json.loads(classifications.annotations[i])
    for j in json_annotation:
        tasks.append(j["task"])
        
unique_tasks = list(np.unique(tasks))
unique_tasks.sort(key=lambda x: int(x.split('T')[1]))
unique_tasks
```
